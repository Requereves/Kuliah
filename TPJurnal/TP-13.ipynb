```python
n, x = map(int, input().split()) 
tiket = list(map(int, input().split())) 

kiri = 0 
kanan = n - 1 
ditemukan = False 

# selama kiri lebih kecil sama dgn kanan 
while kiri <= kanan: 
    # nilai tengah = kiri + kanan dibagi 2
    tengah = (kiri + kanan) // 2 

    # jika tiket idx tengah == x
    if tiket[tengah] == x:
        #  maka print ini
         print("tiket ada di idx", tengah) 
        #  valid
         ditemukan = True 
         break 
    # jika x nilainya lebih kecil dari tiket idx tengah
    elif x < tiket[tengah]:
        #  tengah dikurang -1
         kanan = tengah - 1 
    # tengah + 1
    else: kiri = tengah + 1 
# kalo ga ditemukan
if ditemukan == False: 
    print("Palsu")
```

```python
n, x = map(int, input().split())
nim = list(map(int, input().split()))
ditemukan = False

for i in range(n):
    #  kalo nim idx i == x
     if nim[i] == x:
        # print ini
        print("Ditemukan pada indeks", i)
        # ganti jadi true
        ditemukan = True 
        break 
    # klao ga nemu 
if ditemukan == False: 
    print("Belum Mengumpulkan")
```
