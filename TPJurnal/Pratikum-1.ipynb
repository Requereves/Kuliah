```python
def total(sensor):

    # fungsi rekursif menjumlahkan seluruh data
    def jumlah(sensor):
        # jika panjang sensor == 0 maka returnnya 0
        if len(sensor) == 0:
            return 0
        # jika panjang sensornya ga == 0 maka ngambil value 0 dari array sensor dan membaca ulang dari 1
        return sensor[0] + jumlah(sensor[1:])
    
    # fungsi rekursif menghitung total selisih kuadrat
    def kuadrat(sensor, rata):
        # jika panjang sensor == 0 maka return 0
        if len(sensor) == 0:
            return 0
        # mengambil index 0 dari array sensor lalu di kurang dan dipangkat 2
        return (sensor[0] - rata)**2 + kuadrat(sensor[1:], rata)
    
    # menyimpan value dari fungsi jumlah
    hasil = jumlah(sensor)
    # hitung panjang sensor
    n = len(sensor)
    rata = hasil / n

    var = kuadrat(sensor, rata)
    varians = var / n
    
    return hasil, rata, var, varians

sensor = [22,24,26,20,28]
hasil, rata, var, variansi = (total(sensor))

print("Total Massa Suhu:", hasil)
print("Rata-rata Suhu:", rata)
print("Total Selisih Kuadrat:", var)
print("Variansi Akhir:", variansi)
```

```python
# array kosong buat nyimpen datanya nanti
armada = [] 

# pendaftaran
def daftar_pesawat():
    # menentukan panjang datanya sebanyak 50
    if len(armada) >= 50:
        print("armada penuh")
        return
    
    # membuat input untuk nama kapasitas dan jarak
    nama = input("nama pesawat: ")
    kapasitas = int(input("kapasitas kargo: "))
    jarak = int(input("jarak tempuh: "))
    
    # datanya di simpen disini, dalam bentuk array
    pesawat = [nama, kapasitas, jarak]
    # ini buat masukkin datanya ke array
    armada.append(pesawat)
    
    print("Pesawat ditambahkan")

# ngitung energi dengan 2 * jarak + kapasitas
def hitung_energi(jarak, kapasitas):
    energi = (2 * jarak) + kapasitas
    return energi

# hyperdrive
def kode_warp(warp):
    # jika warpny == 1 maka dia dapet return nilainy 10
    if warp == 1:
        return 10
    # sedangkan jika tidak maka returnnya di kali 2 dengan warp - 1 dan ditambah 5
    else:
        return 2 * kode_warp(warp - 1) + 5
        

# minimal 2 pesawat
for i in range(2):
    print(f"Pesawat ke-{i+1}")
    daftar_pesawat()

# manifes
for pesawat in armada:
    # manggil semua datanya sesuai value arraynya
    nama = pesawat[0]
    kapasitas = pesawat[1]
    jarak = pesawat[2]
    
    energi = hitung_energi(jarak, kapasitas)
    
    print(f"Pesawat: {nama}")
    print(f"Total Energi: {energi}")
    print("")

# inputan warp
warp = int(input("warp berapa: "))
hasil = kode_warp(warp)
print(f"Kalibrasinya: {hasil}")
```

```python
def faktorial(n):
    hasil = 1
    for i in range(1, n+1):
        hasil = hasil * i
    return hasil


def permutasi(a, b):
    return faktorial(a) // faktorial(a-b)


x = int(input("Masukkan x: "))
y = int(input("Masukkan y: "))

fx = faktorial(x)
fy = faktorial(y)

if x >= y:
    p = permutasi(x, y)
else:
    p = permutasi(y, x)

print("x! =", fx)
print("y! =", fy)
print("Permutasi =", p)
```

```python
def digit(n):

    # inisiasi nilai hasil adalah 0
    hasil = 0
    # membuat variabel angka untuk menyimpan nilai dari n
    angka = n

    # menggunakan while karena buat ngecek dulu kondisinya sebelum melakukan perulangan
    while angka > 0:
        
        digit = angka % 10
        hasil = hasil * 10 + digit
        angka = angka // 10

    # inisiasi nilai awal pangkat adalah 1
    pangkat = 1
    # inisiasi nilai awal total adalah 0
    total = 0

    # menggunakan while buat ngecek dulu kondisinya 
    while hasil > 0:
        digit = hasil % 10
        total = total + digit ** pangkat
        hasil = hasil // 10
        pangkat = pangkat + 1

    return total

# membuat input untuk n
n = int(input("masukkan angka: "))
# print
print(digit(n))
```

```python
def number(n):
    # inisiasi awal nilai hasil = 0
    hasil = 0

    # melakukan pembalikan angka menggunakan while
    while n > 0:
        angka = n % 10
        hasil = hasil * 10 + angka
        n = n // 10

    return hasil

# input dan juga print hasil
n = int(input("masukkan angka: "))
print(number(n))
```

```python
def special(n):

    # inisiasi untuk setiap variabel
    angka = n
    jumlah = 0
    total = 0

    # melakukan perulangan
    while angka > 0:
        # mengambil digit terakhir lalu memasukkan nilainy ke variabel jumlah
        jumlah += angka % 10
        # menghapus digit terakhir
        angka = angka // 10
        # untuk menambahkan jumlah digit
        total = total + 1

    # mencari hasil dengan mengkalikan jumlah dan total
    hasil = jumlah * total

    # mengecek apakah nilai dari hasil itu sama dengan nilai dari n?
    if hasil == n:
        return True
    else:
        return False

# input dan juga output
n = int(input("Masukkan angka: "))
print(special(n))
```

```python
def digit(n):

    # inisiasi nilai buat variabel angka dan total
    total = 0
    angka = n

    # membalik angka agar digit dibaca dari depan
    while angka > 0:
        digit = angka % 10
        total = total * 10 + digit
        angka = angka // 10

    # inisiasi lagi nilai buat vaariabel hasil dan op
    hasil = 0
    # 1 untuk + 
    op = 1   

    while total > 0:
        digit = total % 10
        hasil += op * digit
         # mengganti + menjadi - dan sebaliknya
        op = -op       
        total = total // 10

    return hasil


n = int(input("masukkan angka: "))
print(digit(n))
```
