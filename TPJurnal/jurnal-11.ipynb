```python
data_nilai = []

#fungsi buat input user
def input_user(n):
    for i in range(n):
        masukan_nilai = int(input())
        data_nilai.append(masukan_nilai)
        
# fungsi buat cari nilai min
def minimumNilai(t, n):
    min = t[0]
    k = 1
    
    #perulangan buat ngecek min
    while k < n:
        if min > t[k]:
            min = t[k]
            k = k + 1
    return min
# fungsi buat nilai maks
def maksimumNilai(t, n):
    maks = t[0]
    k = 1
#perulangan buat ngecek maks
    while k < n:
        if maks < t[k]:
            maks = t[k]
            k = k + 1
    return maks
#buat rata rata
def avgNilai(data):
    total = 0
    for angka in data:
        total = total + angka
        hasil = total / len(data)
        return hasil
    
j_data = int(input())
input_user(j_data)

maks = maksimumNilai(data_nilai, j_data)
min = minimumNilai(data_nilai, j_data)
rataanNilai = avgNilai(data_nilai)

print(f'Nilai Maksimum: {maks}')
print(f'Nilai Minimum: {min} ')
print(f'Rata-rata: {rataanNilai:.2f}')
```
