```python
n, lclu, t_hujan = input().split()
n = int(n)
t_hujan = float(t_hujan)

hasil = []

# inputan buat data obs kategori sama hujan
for i in range(n):
    obs, kategori, hujan = input().split()
    hujan = float(hujan)
    # kalo ktgore ny == lclu sama hujan lebih besar dari target
    if kategori == lclu and hujan >= t_hujan:
        hasil.append(obs)

if len(hasil) > 0:
    print("Ditemukan", len(hasil), "area kritis:")

    for i in range(len(hasil)):
        print(hasil[i], end=" ")
else:
    print("Aman. Tidak ada anomali.")
```

```python
n, target = map(int, input().split())
data = list(map(int, input().split()))

awal = -1
akhir = -1

# Binary Search buat nyari idx pertama
left = 0
right = n - 1

while left <= right:
    mid = (left + right) // 2

    if data[mid] == target:
        awal = mid
        right = mid - 1
    elif data[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

# Binary Search buat nyari idx terakhir
left = 0
right = n - 1

while left <= right:
    mid = (left + right) // 2

    if data[mid] == target:
        akhir = mid
        left = mid + 1
    elif data[mid] < target:
        left = mid + 1
    else:
        right = mid - 1

if awal == -1:
    print("Data Play_ID tidak ditemukan dalam pipeline.")
else:
    jumlah = akhir - awal + 1
    print("Target ditemukan dari indeks", awal,
          "sampai", akhir + ".",
          "Total frame:", jumlah)   
```
