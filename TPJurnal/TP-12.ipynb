```python
n = int(input())
nilai = list(map(int, input().split()))

# Insertion Sort
for i in range(1, n):
    k = nilai[i]
    j = i - 1 
    # Bnadingin nilai J sama k
    while j >= 0 and nilai[j] > k:
        nilai[j + 1] = nilai[j]
        j -= 1

    nilai[j + 1] = k

for i in range(n):
    print(nilai[i], end=" ")
```

```python
n = int(input())
buku = list(map(int, input().split()))


# Selection sort
for i in range(n - 1):
    idx = i
    # bandingin nilai J sama nilai idx
    for j in range(i + 1, n):
        if buku[j] < buku[idx]:
            # tuker nilai idx jadi nilai j kalo j lebih kecil dari nilai idx
            idx = j

    temp = buku[i]
    buku[i] = buku[idx]
    buku[idx] = temp

for i in range(n):
    print(buku[i], end=" ")
```
